```python
import numpy as np
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
import dlib
from matplotlib import pyplot as plt
```

```python
# original image
img = cv2.imread('bird2.jpg')
print('original image')
plt.imshow(img)
```

```python
# applying YOLOv3
img_color_cvt = cv2.cvtColor(img,cv2.COLOR_BGR2RGB )
bbox, label, conf = cv.detect_common_objects(img_color_cvt)
out = draw_bbox(img, bbox, label, conf)
plt.imshow(out)
  
```

```python
# applying mask

mask = np.zeros(np.array(bbox[0][2:]), np.uint8) 
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
```

```python
bbox
```

```python
# use only for fixed size bounding box 
h = bbox[0][2] 
w = bbox[0][3]
```

```python
# pass the bounding box to grabcut algorithm 
rect = (10, 10, h, w)
rect
```

```python
# croping the image as per YOLOv3 bounding box
crop_image=  img_color_cvt[bbox[0][1] -30 : bbox[0][3]+30 ,
                 bbox[0][0] -30 : bbox[0][2] +30]
plt.imshow(crop_image)
```

```python
# this modifies mask 
mask, bgdModel, fgdModel = cv2.grabCut(crop_image, mask,rect, bgdModel,fgdModel,6,cv2.GC_INIT_WITH_RECT)
```

```python
# If mask==2 or mask== 1, mask 2 get 0, otherwise it gets 1 as 'uint8 
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
mask2
```

```python
# adding additional dimension for rgb to the mask, by default it gets # multiply it with input image to get the segmented image
img_cut=crop_image*mask2[:,:,np.newaxis]
img_cut
```

```python
plt.imshow(img_cut)
```

```python
plt.subplot(131),plt.imshow(img_color_cvt)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132),plt.imshow(out)
plt.title('Appying YOLOv3'), plt.xticks([]), plt.yticks([])

plt.subplot(133),plt.imshow(img_cut)
plt.title('GrabCut'), plt.xticks([]), plt.yticks([])
plt.show()
```
